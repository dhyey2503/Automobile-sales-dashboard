```python
import dash
from dash import dcc, html, Input, Output, State
import dash_bootstrap_components as dbc
import pandas as pd
import plotly.express as px
import plotly.io as pio
from io import StringIO
import base64
import datetime

# Load CSV data
data = pd.read_csv("historical_automobile_sales_updated.csv")
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')
data.dropna(subset=['Year'], inplace=True)
data['Year'] = data['Year'].astype(int)

# Generate dynamic lists
year_list = sorted(data['Year'].unique())
city_list = sorted(data['City'].dropna().unique()) if 'City' in data.columns else []
vehicle_list = sorted(data['Vehicle_Type'].dropna().unique())

# Initialize app
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
app.title = "Automobile Statistics Dashboard"

# App layout
app.layout = dbc.Container([
    dbc.Row([
        dbc.Col(html.H1("Automobile Sales Dashboard", className="text-center mb-4"), width=12)
    ]),

    dbc.Row([
        dbc.Col(dbc.RadioItems(
            id='theme-toggle',
            options=[
                {'label': 'Light', 'value': 'plotly'},
                {'label': 'Dark', 'value': 'plotly_dark'}
            ],
            value='plotly',
            inline=True,
            className='mb-3'
        ), width=6)
    ]),

    dbc.Tabs([
        dbc.Tab(label="Yearly Statistics", tab_id="yearly"),
        dbc.Tab(label="Recession Period Statistics", tab_id="recession")
    ], id='tabs', active_tab='yearly'),

    html.Br(),

    dbc.Row([
        dbc.Col(dcc.Dropdown(
            id='select-year',
            options=[{'label': i, 'value': i} for i in year_list],
            value=year_list[-1],
            placeholder='Select Year'
        ), width=3),

        dbc.Col(dcc.Dropdown(
            id='city-filter',
            options=[{'label': i, 'value': i} for i in city_list],
            multi=True,
            placeholder='Filter by City'
        ), width=3),

        dbc.Col(dcc.Dropdown(
            id='vehicle-filter',
            options=[{'label': i, 'value': i} for i in vehicle_list],
            multi=True,
            placeholder='Filter by Vehicle Type'
        ), width=3),

        dbc.Col(dbc.Button("Export CSV", id='export-csv', color='primary', className='me-2'), width='auto'),
        dbc.Col(dbc.Button("Download Image", id='export-img', color='secondary'), width='auto')
    ], className='mb-4'),

    html.Div(id='output-container')
], fluid=True)

# Callbacks
@app.callback(
    Output('output-container', 'children'),
    [
        Input('tabs', 'active_tab'),
        Input('select-year', 'value'),
        Input('theme-toggle', 'value'),
        Input('city-filter', 'value'),
        Input('vehicle-filter', 'value')
    ]
)
def render_content(tab, year, theme, city, vehicle):
    dff = data.copy()
    if city:
        dff = dff[dff['City'].isin(city)]
    if vehicle:
        dff = dff[dff['Vehicle_Type'].isin(vehicle)]

    graphs = []

    if tab == 'recession':
        rec_data = dff[dff['Recession'] == 1]
        g1 = px.line(rec_data.groupby('Year')['Automobile_Sales'].mean().reset_index(), x='Year', y='Automobile_Sales', title='Avg Sales during Recession', template=theme)
        g2 = px.bar(rec_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index(), x='Vehicle_Type', y='Automobile_Sales', title='Sales by Vehicle Type', template=theme)
        g3 = px.pie(rec_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum().reset_index(), values='Advertising_Expenditure', names='Vehicle_Type', title='Ad Spend Share', template=theme)
        g4 = px.bar(rec_data.groupby(['Vehicle_Type', 'unemployment_rate'])['Automobile_Sales'].mean().reset_index(), x='Vehicle_Type', y='Automobile_Sales', color='unemployment_rate', title='Unemployment vs Sales', template=theme)
        graphs = [g1, g2, g3, g4]

    elif tab == 'yearly':
        yearly = dff[dff['Year'] == year]
        g1 = px.line(dff.groupby('Year')['Automobile_Sales'].mean().reset_index(), x='Year', y='Automobile_Sales', title='Yearly Sales Trend', template=theme)
        g2 = px.line(yearly.groupby('Month')['Automobile_Sales'].sum().reset_index(), x='Month', y='Automobile_Sales', title=f'Monthly Sales in {year}', template=theme)
        g3 = px.bar(yearly.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index(), x='Vehicle_Type', y='Automobile_Sales', title=f'Sales by Vehicle Type in {year}', template=theme)
        g4 = px.pie(yearly.groupby('Vehicle_Type')['Advertising_Expenditure'].sum().reset_index(), values='Advertising_Expenditure', names='Vehicle_Type', title=f'Ad Spend in {year}', template=theme)
        graphs = [g1, g2, g3, g4]

    return dbc.Row([
        dbc.Col(dcc.Graph(id=f'graph-{i}', figure=fig), width=6) for i, fig in enumerate(graphs)
    ], className='gy-4')

@app.callback(
    Output('export-csv', 'href', allow_duplicate=True),
    Output('export-csv', 'download', allow_duplicate=True),
    Input('export-csv', 'n_clicks'),
    State('select-year', 'value'),
    State('city-filter', 'value'),
    State('vehicle-filter', 'value'),
    prevent_initial_call='initial_duplicate')
def download_csv(n, year, city, vehicle):
    dff = data.copy()
    if city:
        dff = dff[dff['City'].isin(city)]
    if vehicle:
        dff = dff[dff['Vehicle_Type'].isin(vehicle)]
    dff = dff[dff['Year'] == year]
    csv_str = dff.to_csv(index=False)
    csv_b64 = base64.b64encode(csv_str.encode()).decode()
    filename = f"automobile_data_{year}.csv"
    return f"data:text/csv;base64,{csv_b64}", filename

@app.callback(
    Output('export-img', 'href', allow_duplicate=True),
    Output('export-img', 'download', allow_duplicate=True),
    Input('export-img', 'n_clicks'),
    State('select-year', 'value'),
    prevent_initial_call='initial_duplicate')
def download_image(n_clicks, year):
    fig = px.line(data[data['Year'] == year].groupby('Month')['Automobile_Sales'].sum().reset_index(), x='Month', y='Automobile_Sales')
    fig_bytes = fig.to_image(format="png")
    b64_img = base64.b64encode(fig_bytes).decode()
    filename = f"sales_{year}.png"
    return f"data:image/png;base64,{b64_img}", filename

if __name__ == '__main__':
    app.run(debug=True)
```
